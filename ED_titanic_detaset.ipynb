```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("Titanic_clean_data.csv")
```

```python
df
```

```python
df.info()
```

```python
#pie chart for "sex"
explode = [0, 0.2] 
sex_count = df["Sex"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(sex_count ,labels=sex_count.index ,autopct="%1.1f%%",colors=["skyblue","yellow"],wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--'},explode=explode,shadow=True, startangle =100)
plt.title("gender distribution",color="black",fontsize=19)
# autopct ="%1.1f%% : used to represent the numbers on the plot we can extend the decimal value by ".1f%%",".2%%"
#explode = use to slice the pie chart  & firstly we have to give assign a value to explode like-explode = [0, 0.2] 
# in the pie chart we can add the design using wedgeprops
```
