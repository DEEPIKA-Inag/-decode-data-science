```python
import matplotlib.pyplot as plt
```

```python
x =[1,2,3,4,5,5,6,7,8,9,10 ]
y1 =[i * 2 for i in x]
y2 = [i ** 2 for i in x]
```

```python
plt.subplot(1,2,1) # row , column, plot number
plt.plot(x,y1)
plt.title("double of x")
plt.xlabel("number")
plt.ylabel("ioubel of x")

plt.subplot(1, 2,2)
plt.plot(x ,y2)
plt.title("square of x")
plt.xlabel("number")
plt.ylabel("square of nubers")
plt.tight_layout() # to fix overlaping
```

```python
x =[1,2,3,4,5,5,6,7,8,9,10 ]
y1 =[i * 2 for i in x]
y2 = [i ** 2 for i in x]
y3 =[ i ** 0.5 for i in x]
y4 =[ 10 -i for i in x]
```

```python
plt.figure(figsize=(8,6))
plt.subplot(2,2,1) # row , column, plot number
plt.plot(x,y1)
plt.title("double of x")
plt.xlabel("number")
plt.ylabel("doubel of x")

plt.subplot(2, 2,2)
plt.plot(x ,y2)
plt.title("square of x")
plt.xlabel("number")
plt.ylabel("square of nubers")
plt.tight_layout() # to fix overlaping


plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("squrt(x)")


plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("10-x")
plt.tight_layout()
plt.show()

```

```python
fig , axs =plt.subplots(1,2,figsize=(10,4))
axs[0].plot(x,y1) # axs work on individual plots, and fig work on whole figure
axs[0].set_title("x*2")

axs[1].plot(x,y2)
axs[1].set_title("x**2")

fig.suptitle("my comparision",fontsize=20)
```
