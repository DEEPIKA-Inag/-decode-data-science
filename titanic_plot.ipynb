```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Titanic_clean_data.csv")
```

```python
df
```

```python
df.head()
```

```python
#pie chart for "sex"
explode = [0, 0.1] 
sex_count = df["Sex"].value_counts()
plt.figure(figsize=(5,5))
plt.pie(sex_count ,labels=sex_count.index ,autopct="%1.1f%%",colors=["pink","grey"],wedgeprops={'edgecolor': 'pink', 'linewidth': 2, 'linestyle': '--'},explode=explode,shadow=True, startangle =100)

plt.title("gender distribution")
```

```python
# pie chart for SibSp
sibsip = df["SibSp"].value_counts()
plt.figure(figsize=(4,4))
plt.pie(sibsip ,labels=sibsip ,colors=["pink","black","orange","green","red","skyblue","white"],autopct="%1.1f%%",textprops={'fontsize': 8})
plt.title("sibsp")
plt.legend()
plt.tight_layout()
# NOTE:  Donâ€™t use pie charts for columns like SibSp, Pclass, Fare, Age, etc. if they have many values.
 # Use bar plots, count plots, or histograms for clearer, more effective visualizations.
```

```python
sns.lineplot(data=df,x="Age",y="SibSp")
plt.title("plot b/w sibsp and age")
plt.grid(True)
```

```python
sns.boxplot(data=df,x="Age",y="SibSp")
```

```python
sns.scatterplot(data=df,x="Age",y="SibSp")
```

```python
survived_count =df["Survived"].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.pie(survived_count,autopct="%1.1f%%",labels=survived_count.index,colors=["pink","red","orange"],textprops={"fontsize":20,"color":"orange"},wedgeprops={"edgecolor":"black","linewidth":1})
plt.title("female VS male")
# where 0=male, 1= female
```

```python
sns.countplot(data=df ,x="Survived")
plt.xticks([0,1],["not survived","survived"])
plt.title("plot of survived")
plt.grid(True)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(29, 19))

# Age vs sex
axs[0].scatter(df['Age'], df['Sex'], alpha=0.5, color='teal')
axs[0].set_title("Age vs sex")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Sex")

# Gender Count
sns.countplot(data=df, x='Sex', ax=axs[1])
axs[1].set_title("Gender Count")

# Survival Count
sns.countplot(data=df, x='Survived', ax=axs[2])
axs[2].set_title("Survival Count")
axs[2].set_xticklabels(['Not Survived', 'Survived'])

plt.tight_layout()
plt.show()
```

```python
sns.histplot(data=df, x='Age', bins=20, kde=True, color='pink')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
# here you can use kde=True to add a smooth line (Kernel Density Estimation)
```

```python
sns.histplot(data=df, x='Age', bins=20, kde=False, color='pink') ## kde=false
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python

# Compute correlation matrix
corr = df.corr(numeric_only=True)

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=4.5)

plt.title("Correlation Heatmap")
plt.show()
# Create a pivot table
pivot = df.pivot_table(index='Sex', columns='Embarked', values='Survived', aggfunc='mean')

# Plot heatmap
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.title("Survival Rate by Sex and Pclass")
plt.show()
```
