```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
```

```python
plt.xlabel("year")
```

```python
plt.ylabel("runs")
```

```python
plt.plot(years,runs)
plt.xlabel("year")
plt.ylabel("runs")
plt.title("sachin ,yearly runs!")
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label="Virat Kohli")
plt.plot(years, sehwag, label="Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

```python
plt.plot(years, kohli, 'ro--', label="Kohli")  # red circles with dashed lines
plt.plot(years, sehwag, 'g^:', label="Sehwag")  # green triangles dotted
plt.legend()
```

```python
plt.style.use('ggplot')
```

```python
plt.plot(years, kohli, color ="black", linestyle= '--',linewidth= 6, label="Kohli")  # red circles with dashed lines
plt.plot(years, sehwag, color ="green", linestyle = "-.",label="Sehwag")# green triangles dotted
plt.plot(years,runs,color ="blue",label ="sachin")
plt.grid(True)
plt.tight_layout()
plt.legend()
```
