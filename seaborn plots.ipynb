```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names())
```

```python
tips =sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.info()
```

```python
sns.barplot(x="total_bill", y="tip" , data=tips)
plt.title("Line plot")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
sns.boxplot(x="total_bill", y="tip" , data=tips)
plt.title("Line plot")
plt.xlabel("X")
plt.ylabel("Y")
```
