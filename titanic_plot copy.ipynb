```python
sns.countplot(data=df ,x="Survived",palette=["lightgreen","skyblue"])
plt.xticks([0,1],["not survived","survived"])
plt.title("plot of survived")
plt.legend()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(9, 9))

# Age vs sex
axs[0].scatter(df['Age'], df['Sex'], alpha=0.5, color='teal')
axs[0].set_title("Age vs sex")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Sex")

# Gender Count
sns.countplot(data=df, x='Sex', ax=axs[1])
axs[1].set_title("Gender Count")

# Survival Count
sns.countplot(data=df, x='Survived', ax=axs[2])
axs[2].set_title("Survival Count")
axs[2].set_xticklabels(['Not Survived', 'Survived'])
#survival rate by embarkation port
ort_survival_rate = df.groupby('Embarked')['Survived'].mean().sort_values(ascending=False).reset_index()

# Calculate survival rate per port
port_survival_rate = df.groupby('Embarked')['Survived'].mean().sort_values(ascending=False).reset_index()

# Plot
plt.figure(figsize=(6,4))
sns.barplot(x='Embarked', y='Survived', data=port_survival_rate, palette='Blues')

plt.title('Survival Rate by Embarkation Port')
plt.ylabel('Survival Rate')
plt.xlabel('Embarked Port')
plt.grid(True, axis='y', linestyle=':', alpha=0.7)
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

```python
sns.histplot(data=df, x='Age', bins=20, kde=True, color="skyblue",edgecolor="black")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
# here you can use kde=True to add a smooth line (Kernel Density Estimation)
```
