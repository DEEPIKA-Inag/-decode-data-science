```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# here we have to use pandas for catch ,read or analysis the dataS
# Matplotlib : shows the relation between data by polts
# Seaborn : shows the advance plots and easy to implement
```

```python
df = pd.read_csv("Titanic_clean_data.csv")
```

```python
df.head()
```

```python
#pie chart for "sex"
explode = [0, 0.2] 
sex_count = df["Sex"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(sex_count ,labels=sex_count.index ,autopct="%1.1f%%",colors=["skyblue","yellow"],wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--'},explode=explode,shadow=True, startangle =100)
plt.title("gender distribution",color="black",fontsize=19)
# autopct ="%1.1f%% : used to represent the numbers on the plot we can extend the decimal value by ".1f%%",".2%%"
#explode = use to slice the pie chart  & firstly we have to give assign a value to explode like-explode = [0, 0.2] 
# in the pie chart we can add the design using wedgeprops
```

```python
# in this above code , shows the gender distribution,it shows the percentage of females and males
```

```python
# pie chart for SibSp"].value_counts()
sibsip = df["SibSp"].value_counts().sort_index()
explode =[0,0,0,0.9,1,0.9,0.9]
plt.figure(figsize=(8,8))
plt.pie(sibsip ,labels=sibsip ,colors=["yellow","black","orange","green","red","skyblue","white"],autopct="%1.1f%%",textprops={'fontsize': 8},explode=explode,shadow=True)
plt.title("sibsp",fontsize=19)
plt.legend()
plt.tight_layout()
```

```python
# We should not take too many values in the pie chart as that will not give a good representation of the data,we should represent only two or three values through pie chart
#EDA:
#This pie chart represents the distribution of passengers on the Titanic based on the number of siblings or spouses they had aboard the ship. The data comes from the SibSp column in the Titanic dataset.
#The majority of passengers likely had 0 siblings/spouses aboard, which will appear as the largest slice.

#Smaller slices represent passengers with 1, 2, 3, etc., relatives.

#The exploded slices (3, 4, 5, 8) might be rare cases but are visually highlighted to draw attention.

#Helps identify how family size aboard varied among passengers.

```

```python
sns.lineplot(data=df,x="Age",y="SibSp",color="blue",linestyle=":",linewidth=2)
plt.title("plot b/w sibsp and age",fontsize=18,color="blue")
plt.xlabel("Age",color="blue", fontsize=15)
plt.ylabel("Sibsp",color="blue",fontsize=15)
plt.grid(True,linestyle="--",color="pink")
plt.gca().set_facecolor("black")
# gca().set_facecolor("color") is used to add the background-color
```

```python
# The line plot reveals how the number of siblings/spouses aboard (SibSp) varies with passenger age.
```

```python
sns.scatterplot(data=df,x="Age",y="SibSp")
```

```python
# This scatter plot displays the relationship between a passenger's age and the number of siblings or spouses (SibSp) they had aboard.
```

```python
survived_count =df["Survived"].value_counts()
```

```python
sns.countplot(data=df ,x="Survived",palette=["lightgreen","skyblue"])
plt.xticks([0,1],["not survived","survived"])
plt.title("plot of survived")
plt.legend()
```

```python
#in  above plot : comparing the height of the bars, we can easily understand the survival distribution among all passengers.
# where, 0 as "not survived" and 1 as "survived" 
```

```python
sns.histplot(data=df, x='Age', bins=20, kde=True, color="skyblue",edgecolor="black")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
# here you can use kde=True to add a smooth line (Kernel Density Estimation)
```

```python
#This histogram displays the age distribution of passengers aboard the Titanic. Each bar represents the count of passengers within a specific age range, giving a clear view of which age groups were most common.
```

```python
# note-here you can use kde=False to remove a smooth line (Kernel Density Estimation)
```

```python
data = pd.crosstab(df["Sex"],df["Survived"])
sns.heatmap(data,annot=True ,cmap="cool")
plt.title("Survival Count by Gender")
```

```python
#This heatmap visualizes the survival count of passengers based on gender. It shows the actual number of males and females who survived (Survived = 1) and who did not survive (Survived = 0).
#This code helps you visually analyze survival rates by gender using a heatmap:

#Males had lower survival rates.

#Females had higher survival, possibly due to rescue priorities.

#Itâ€™s a compact way to compare categorical variables (Sex vs Survived) using both numbers and colors.
```

```python
sns.histplot(data=df, x='Age', hue='Survived', kde=True)
plt.title("Age Distribution by Survival Outcome")
```

```python
#it allows us to compare how age relates to survival.
#With kde=True, smooth density curves are overlaid for both survived and not survived groups, helping to visualize the age trends more clearly.
```

```python
sns.set(style="whitegrid")
sns.countplot(x='Embarked', data=df, hue='Embarked', palette=["skyblue","yellow","pink"], legend=False)
plt.title("number of passengers")
plt.xlabel('embarkation port')
plt.ylabel('number of Passengers')
plt.grid(True)
plt.legend()

```
